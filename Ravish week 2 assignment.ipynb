Q2. What does inplace=True do? Why is the pandas team discouraging its use in newer versions? Give
one example where you'd avoid it.

A.

inplace=True modifies the original DataFrame or Series directly instead of returning a new object.

It discouraged because:-

1)It can make code harder to read and debug.

2)It does not always improve performance or memory usage.

3)A consistent "return a new object" approach is easier to understand and maintain, so newer pandas versions encourage avoiding inplace=True.

```python
# Avoid
df.drop('Age', axis=1, inplace=True)

# Preferred
df = df.drop('Age', axis=1)
```

Q4. What does pd.to_numeric(col, errors='coerce') do that plain pd.to_numeric(col) does NOT? Which
one would you use on real-world messy data, and why?

A.

**pd.to_numeric(col)** converts the values in a column to numeric data types. If it encounters an invalid value (such as "abc" or "N/A"), it raises an error and stops execution.

**pd.to_numeric(col, errors='coerce')** also converts values to numeric, but any invalid or non-numeric values are automatically replaced with NaN instead of raising an error.

Example:

```python
import pandas as pd

col = pd.Series(["10", "25", "abc", "40"])

pd.to_numeric(col, errors='coerce')
```

On real-world messy data we can use **pd.to_numeric(col, errors='coerce')** because real-world datasets often contain invalid values such as "N/A", "unknown", or text mixed with numbers. Converting these values to NaN allows the data to be cleaned later using methods like .fillna() or .dropna() without causing the program to fail.

Q5. Explain .loc vs .iloc with one example each. Then describe a case where df.loc[0] and df.iloc[0]
return different rows.

A.

**.loc** is label-based indexing. It selects rows and columns using index labels.

**.iloc** is integer position-based indexing. It selects rows and columns using row and column positions.

Example of **.loc**

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"]
}, index=[101, 102, 103])

print(df.loc[102])
```

Example of **.iloc**

```python
print(df.iloc[1])
```

Thus, they return different rows because:

**.loc** → Label-based indexing.

**.iloc** → Position-based indexing.

They return different rows when the DataFrame index labels are not in the same order as the row positions.

Q7. What is the SettingWithCopyWarning? Give one example of code that triggers it and one way to fix
it. (Hint: the chained-assignment pattern.)

A.


SettingWithCopyWarning is a warning in pandas that appears when you try to modify a slice (copy) of a DataFrame instead of the original DataFrame. This usually happens because of chained assignment, and the changes may not be applied as expected.

Example that triggers the warning:

```python
import pandas as pd

df[df["Age"] > 18]["Status"] = "Adult"
```

This uses chained assignment:

df[df["Age"] > 18] creates a filtered DataFrame.

["Status"] = "Adult" tries to modify that filtered object.

This can produce a SettingWithCopyWarning.

Correct way to fix it

Use .loc to modify the original DataFrame directly:

```python
df.loc[df["Age"] > 18, "Status"] = "Adult"
```

This safely updates the original DataFrame without generating the warning.

Q17.After loading the raw CSV (before cleaning), answer:

(a) How many rows and columns?

(b) How many missing values are in each column?

(c) How many exact duplicate rows are present?



```python
import pandas as pd

df = pd.read_csv("indian_startups_fundingDADS.csv")

# (a)
print("Rows, Columns:", df.shape)

# (b)
print("\nMissing values:")
print(df.isna().sum())

# (c)
print("\nDuplicate rows:", df.duplicated().sum())
```





> **Cleaning Cell:**



```python
import pandas as pd
import numpy as np

df = pd.read_csv("indian_startups_fundingDADS.csv")

# 1. Rename columns
df.columns = [
    "startup_name",
    "city",
    "industry",
    "funding_stage",
    "amount",
    "funding_date",
    "investors",
    "founded_year"
]

# 2. Clean startup_name & city
df["startup_name"] = df["startup_name"].str.strip().str.title()
df["city"] = df["city"].str.strip().str.title()

# 3. Standardise city aliases
city_map = {
    "Bangalore": "Bengaluru",
    "Blr": "Bengaluru",
    "Bombay": "Mumbai",
    "Gurgaon": "Gurugram",
    "Delhi": "New Delhi",
    "Calcutta": "Kolkata",
    "Madras": "Chennai"
}

df["city"] = df["city"].replace(city_map)

# 4. Standardise industry
industry_map = {
    "Fin Tech": "Fintech",
    "Fintech": "Fintech",
    "Ed Tech": "Edtech",
    "Edtech": "Edtech",
    "Health Tech": "HealthTech",
    "Healthtech": "HealthTech",
    "Health Tech ": "HealthTech",
    "E-Commerce": "Ecommerce",
    "Ecommerce": "Ecommerce",
    "Saas": "SaaS"
}

df["industry"] = df["industry"].str.strip().replace(industry_map)

# Standardise funding stage
stage_map = {
    "Seed Round": "Seed",
    "Seed Funding": "Seed",
    "Series A Round": "Series A",
    "Series B Round": "Series B",
    "Series C Round": "Series C",
    "Pre-Series A": "Pre-Series A"
}

df["funding_stage"] = df["funding_stage"].str.strip().replace(stage_map)

# 5. Clean amount
df["amount"] = (
    df["amount"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("INR", "", regex=False)
    .str.replace("Cr", "", regex=False)
    .str.replace("Crore", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace("Undisclosed", "", regex=False)
    .str.strip()
)

df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

# 6. Date
df["funding_date"] = pd.to_datetime(
    df["funding_date"],
    errors="coerce",
    dayfirst=True
)

# 7. Remove duplicates
df = df.drop_duplicates()
```
